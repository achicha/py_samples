```python
import requests, json


moex_symbols = []

start, end = 0, 1
for i in range(1, 7):
    url = 'http://www.moex.com/iss/apps/rms/rates.jsonp?callback=iss_jsonp_71253585609d51c95ce80372c59f5415f0270181&iss.meta=off&iss.only=rates%2Crates.cursor&sort_column=SECID&sort_order=ASC&currencyid=&collateral=0&security_types=common_share%2Cpreferred_share&board_groups=stock_tplus%2Cstock_ndm_tplus&listname=1%2C2%2C3%2C_&index=&start={}&limit={}&lang=en&_=1480420833558'.format(start * 50, end * 50)
    start, end = end, end + 1
    resp = requests.get(url)
    js = json.loads(resp.content[51:-1].decode('utf-8'))
    for j in js['rates']['data']:
        moex_symbols.append(j[0])
```

```python
not_TQBR = []
symbols_info = {}
for sym in set(moex_symbols):
    url2 = 'http://www.moex.com/iss/engines/stock/markets/shares/boards/TQBR/securities/{}.jsonp?callback=iss_jsonp_aed9dbe8a9179a31df465e87d60d8dac5a606188&iss.meta=off&iss.only=securities%2Cmarketdata&lang=en&_=1480420656461'.format(sym)
    try:
        resp = requests.get(url2)
    except:
        not_TQBR.append(sym)
    
    js = json.loads(resp.content[51:-1].decode('utf-8'))
    symbols_info[sym] = js
```
